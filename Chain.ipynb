```python

from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('Seminar.pdf')
docs = loader.load()
docs
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 200)
chunk_document=text_splitter.split_documents(docs)
chunk_document
```

```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS

db= FAISS.from_documents(chunk_document,OllamaEmbeddings())
db
```

```python
query = "The health of the fetus during pregnancy is important because this is an important period in a woman’s life."
result = db.similarity_search(query)
print(result)
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template("""
Answer the following questions based only on the provided context.
Think step by step before provididng the a detailed answer.
I will tip you $1000 if the user finds the answer helpful.
<context>
{context}   
</context>
Question:{input}                                                                                 
""")
```

```python
from langchain_community.llms import Ollama
llm = Ollama(model="llama2")
llm
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain
doc_chain=create_stuff_documents_chain(llm,prompt)
doc_chain
```

```python
# Retrievers
retriever=db.as_retriever()
retriever
```

```python
# Retriever Chain
from langchain.chains import create_retrieval_chain
retriever_chain = create_retrieval_chain(retriever,doc_chain)
```

```python
response=retriever_chain.invoke({"input":"Fetal ECG extraction"})
response['answer']
```
